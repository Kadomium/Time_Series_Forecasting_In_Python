```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
df = pd.read_csv('Ichijo_cdno1.csv')
df.head()
```

```python
df.tail()
```

```python
df['date_time'] = pd.to_datetime(df['date_time'])
```

```python
df = df.drop_duplicates(subset='date_time', ignore_index=True)
```

```python
df.shape
```

```python
date_range = list(pd.date_range('2020-01-01 00:00:00', '2021-12-31 23:00:00', freq='H'))
print(len(date_range))
```

```python
new_df = pd.DataFrame({'date_time': date_range})
new_df.head()
```

```python
full_df = pd.merge(new_df, df, how='left', on='date_time')
```

```python
full_df.shape
```

```python
full_df.head()
```

```python
full_df.isna().sum()
```

```python
fig, ax = plt.subplots(figsize=(13,6))

ax.plot(full_df.n_main_consumption)
ax.set_xlabel('Date')
ax.set_ylabel('consumption')

fig.autofmt_xdate()
plt.tight_layout()
```

```python
full_df = full_df[35000:].reset_index(drop=True)
```

```python
full_df = full_df.drop(['cdno', 'year', 'month','day','hour'], axis=1)
full_df.shape
```

```python
full_df = full_df.fillna(full_df.groupby(full_df.date_time.dt.hour).transform('median'))
```

```python
full_df.isna().sum()
```

```python
fig, ax = plt.subplots(figsize=(13,6))

ax.plot(full_df.n_main_consumption)
ax.set_xlabel('Date')
ax.set_ylabel('consumption')

fig.autofmt_xdate()
plt.tight_layout()
```
